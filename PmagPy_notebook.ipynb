### Reading data from MagIC format results files (`pmag.magic_read`)

First, the data needs to be imported into the notebook environment. PmagPy has built in functions to deal with data in MagIC database format files. One such function is the `pmag.magic_read` function. In the code block below, this function is used to read in data from a `pmag_results.txt` file. The `pmag.magic_read` function creates two outputs: 

1. the data from the Magic template file in a list of dictionaries 
2. the type of data in the file (e.g. 'pmag_results')

These data were downloaded from the MagIC database (http://earthref.org/MAGIC/9207/) as a .txt file. The data were then unpacked on the command line using the `download_magic.py` program. We will concentrate on importing and using the resulting `pmag_results.txt` file below. The code below relies on the pandas (pd) dataframe structure which is a useful and user friendly way to wrangle data.

```python
#read in the data
data,file_type=pmag.magic_read('Tauxe2004_Data/pmag_results.txt')
#create a pandas dataframe
results = pd.DataFrame(data)
#the data come in as strings so need to be defined as floating point numbers
results.average_dec = results.average_dec.astype(float)
results.average_inc = results.average_inc.astype(float)
results.vgp_lat  = results.vgp_lat .astype(float)
results.vgp_lon  = results.vgp_lon .astype(float)
#display the first 5 rows of the results dataframe
results.head()
```

### Plotting site mean directions

Now that we have imported the data, we can do other things with it such as plot it using `ipmag.plot_DI_mean`.

```python
fignum = 1
plt.figure(num=fignum,figsize=(6,6),dpi=160)
ipmag.plot_net(fignum)
plt.title('Snake River Plain site mean equal area plot')
ipmag.plot_di(results['average_dec'],results['average_inc'])
```

### Calculating and plotting Fisher means from the data

Fisher means for the data can be calculated using the Fisher mean pmag.py function (`pmag.fisher_mean`). In the code below the normal and reversed directions are grouped into lists of [dec,inc] before the means are calculated.

```python
normal_directions = results[results.average_inc>0][['average_dec','average_inc']].values
reverse_directions = results[results.average_inc<=0][['average_dec','average_inc']].values

normal_mean = pmag.fisher_mean(normal_directions)
reverse_mean = pmag.fisher_mean(reverse_directions)
```

The pmag.fisher_mean function returns a dictionary that gives the parameters associated with calculating a Fisher mean. These individual values can be called upon (e.g. `normal_mean['dec']`). A plot can be made of these calculate means along with their $\alpha_{95}$ confidence ellipses using `ipmat.plot_di_mean`.

```python
fignum = 1
plt.figure(num=fignum,figsize=(5,5))
ipmag.plot_net(fignum)
ipmag.plot_di_mean(normal_mean['dec'],normal_mean['inc'],normal_mean['alpha95'],
                  color='b',marker='s',label='mean of normal directions',legend='yes')
ipmag.plot_di_mean(reverse_mean['dec'],reverse_mean['inc'],reverse_mean['alpha95'],
                  color='r',marker='s',label='mean of reverse directions',legend='yes')
```

### Conducting reversal tests on the data

Reversal tests are tests for a common mean between two data sets wherein the antipode vectors from one population are compared to the vectors of another population. The code below conducts the Watson V test and returned the McFadden and McEllhinny (1990) classification and conducts a bootstrap reversal test. The 

```python
ipmag.watson_common_mean(normal_directions,ipmag.flip(reverse_directions),NumSims=1000,plot='yes')
```

```python
ipmag.bootstrap_common_mean(normal_directions,ipmag.flip(reverse_directions),NumSims=1000)
```

### Plotting virtual geomagnetic and mean poles

The normal and reversed data are split into individual data frames in the code below with reversed values being defined for the reversed polarity VGPs so that they can be plotted with the normal VGPs. A `combined_VGP_mean` is calculated using `pmag.fisher_mean`.

```python
Snake_River_N = results.ix[results.vgp_lat>0]
Snake_River_R = results.ix[results.vgp_lat<=0]
Snake_River_R['vgp_lon_rev'] = Snake_River_R['vgp_lon']
Snake_River_R['vgp_lat_rev'] = -Snake_River_R['vgp_lat']

combined_vgps = np.concatenate((Snake_River_N[['vgp_lon','vgp_lat']].values,
                                Snake_River_R[['vgp_lon_rev','vgp_lat_rev']].values))

combined_VGP_mean = pmag.fisher_mean(combined_vgps)
```

To plot poles we first need a map projection. The matplotlib basemap package does a nice job of that, so let's import it below. Note that not all installations of python will have the basemap package so you may need to download it. It is an availible package through Enthought Canopy.

```python
from mpl_toolkits.basemap import Basemap
```

The basemap is highly customizable. A view from space type projection ('ortho') is a nice way to view data on a sphere so let's use that one for this example. We will define an object ('m') that is a map projection and then draw additional things on the map such as continents and lat/long lines. Then the VGPs from the study can be plotted using the `ipmag.plot_vgp` function and the mean pole can be plotted using `ipmag.plot_pole`.

```python
m = Basemap(projection='ortho',lat_0=70,lon_0=230,resolution='c',area_thresh=50000)
plt.figure(figsize=(8, 8))
m.drawcoastlines(linewidth=0.25)
m.fillcontinents(color='bisque',lake_color='white',zorder=1)
m.drawmapboundary(fill_color='white')
m.drawmeridians(np.arange(0,360,30))
m.drawparallels(np.arange(-90,90,30))

ipmag.plot_vgp(m,Snake_River_N['vgp_lon'].tolist(),
               Snake_River_N['vgp_lat'].tolist(),
               color='b',label='normal VGPs')
ipmag.plot_vgp(m,Snake_River_R['vgp_lon_rev'].tolist(),
               Snake_River_R['vgp_lat_rev'].tolist(),
               color='r',label='reversed VGPs')
ipmag.plot_pole(m,combined_VGP_mean['dec'],combined_VGP_mean['inc'],
               combined_VGP_mean['alpha95'],marker='d',label='mean pole')
plt.legend()
plt.title('Snake River Plane VGPs')
plt.show()
```
