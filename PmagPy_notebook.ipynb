Data analysis in Python benefits from an increasingly robust ecosystem of packages for scientific computing and plotting. A tool that has gained increasingly widespread use for conducting and presenting data analysis is the IPython/Jupyter notebook. The IPython project began as a way to bring increased interactivity to data analysis in Python \citep{Perez2007a} and has evolved to include an interactive notebook environment that seeks to enable the full trajectory of scientific computing from initial analysis and visualization onwards to collaboration and through to publication. The project has expanded to enable a reproducible interactive computing environment for many other programming languages (such as R and Julia) and the language-agnostic parts of its architecture have recently been rebranded as Project Jupyter (http://jupyter.org). IPython/Jupyter notebooks allow for executable code, results, text and graphical output to coherently coexist in a single document. With these combined components, they are excellent tools both for conducting data analysis and presenting results. 

One of the major challenges and opportunities in modern science is the communication of complex and varied data analysis in a way that is open and reproducible. The MagIC database plays a significant role as a repository for data and the PmagPy software enables analysis of these data. However, it is difficult to document the many steps that go into data analysis and synthesis. Using PmagPy code within a notebook environment can play a role both as a research tool and a research product that bridges this gap. 

Underlying the PmagPy programs that are accessible through the command line and the GUI interfaces (QuickMagic, demag_GUI, thellier_GUI) are two main function libraries: pmag.py and pmagplotlib.py. These functions can easily be imported and called upon within an IPython/Jupyter notebook. In addition to these functions, we have created a module called IPmag.py that is in active development and contains functions that replicate and extend some of the functionality that is found within the PmagPy command line programs. 

To illustrate some of what is possible in terms of data analysis using PmagPy in a notebook environment, we have created an example notebook that is availible as supplementary material to this paper and also hosted in an open GitHub repository: https://github.com/Swanson-Hysell/2015_Tauxe-et-al_PmagPy_Notebook. This notebook imports data of Snake River Plain lava flows downloaded from the MagIC database (Tauxe et al., 2004), plots the data and conducts some statistical analyses. In the notebook, reversal tests are conducted using both the McFadden and McElhinny (1990) implementation of the Watson (1983) V statistic test for a common mean and the bootstrap reversal test (Tauxe et al., 1991). Conducting these tests using PmagPy in the notebook allows for the underlying code of the test to be availible and for the decisions made in the specific implementation of the test to be transparently presented. For example, the McFadden and McElhinny (1990) reversal test which necessitates a Monte Carlo simulation to determine the critical value of the V statistic. Given that the test relies on Monte Carlo simulation, decisions need to be made as to how many simulations are conducted and every implementation of the test will be unique (although with enough simulations each implementation should yield similar overall results). Publishing a notebook along with a paper allows for the unique implementation of the data analysis used by the authors to be presented and also allows others to reproduce, or revisit in a different way, the results.

### Reading data from MagIC format results files (`pmag.magic_read`)

First, the data needs to be imported into the notebook environment. PmagPy has built in functions to deal with data in MagIC database format files. One such function is the `pmag.magic_read` function. In the code block below, this function is used to read in data from a `pmag_results.txt` file. The `pmag.magic_read` function creates two outputs: 

1. the data from the Magic template file in a list of dictionaries 
2. the type of data in the file (e.g. 'pmag_results')

These data were downloaded from the MagIC database (http://earthref.org/MAGIC/9207/) as a .txt file. The data were then unpacked on the command line using the `download_magic.py` program. We will concentrate on importing and using the resulting `pmag_results.txt` file below. The code below relies on the pandas (pd) dataframe structure which is a useful and user friendly way to wrangle data.

```python
#read in the data
data,file_type=pmag.magic_read('Tauxe2004_Data/pmag_results.txt')
#create a pandas dataframe
results = pd.DataFrame(data)
#the data come in as strings so need to be defined as floating point numbers
results.average_dec = results.average_dec.astype(float)
results.average_inc = results.average_inc.astype(float)
results.vgp_lat  = results.vgp_lat .astype(float)
results.vgp_lon  = results.vgp_lon .astype(float)
#display the first 5 rows of the results dataframe
results.head()
```

### Plotting site mean directions

Now that we have imported the data, we can do other things with it such as plot it using `IPmag.plotDImean`.

```python
fignum = 1
plt.figure(num=fignum,figsize=(6,6),dpi=160)
IPmag.plotNET(fignum)
plt.title('Snake River Plain site mean equal area plot')
IPmag.plotDI(results[['average_dec','average_inc']].values)
```

### Calculating and plotting Fisher means from the data

Fisher means for the data can be calculated using the Fisher mean pmag.py function

```python
#group the normal directions
normal_directions = results[results.average_inc>0][['average_dec','average_inc']].values
#group the reverse directions
reverse_directions = results[results.average_inc<=0][['average_dec','average_inc']].values
reverse_directions_flipped = IPmag.flip(reverse_directions)
combined_directions = np.concatenate((normal_directions,reverse_directions_flipped))

normal_mean = pmag.fisher_mean(normal_directions)
reverse_mean = pmag.fisher_mean(reverse_directions)
combined_mean = pmag.fisher_mean(combined_directions)
combined_mean
```

The pmag.fisher_mean function returns a dictionary that gives the parameters associated with calculating a Fisher mean. These individual values can be called upon (e.g. `combined_mean['dec']`). A plot can be made of these calculate means along with their $\alpha_{95}$ confidence ellipses.

```python
fignum = 1
plt.figure(num=fignum,figsize=(6,6),dpi=160)
IPmag.plotNET(fignum)
IPmag.plotDImean(combined_mean['dec'],combined_mean['inc'],combined_mean['alpha95'],
                  color='g',marker='s',label='mean of combined directions',legend='yes')
IPmag.plotDImean(normal_mean['dec'],normal_mean['inc'],normal_mean['alpha95'],
                  color='b',marker='s',label='mean of normal directions',legend='yes')
IPmag.plotDImean(reverse_mean['dec'],reverse_mean['inc'],reverse_mean['alpha95'],
                  color='r',marker='s',label='mean of reverse directions',legend='yes')
```

### Conducting tests for a common mean on the data

```python
IPmag.WatsonV_common_mean(normal_directions,IPmag.flip(reverse_directions),NumSims=10000,plot='yes')
```

```python
IPmag.bootstrap_common_mean(normal_directions,IPmag.flip(reverse_directions),NumSims=10000)
```

### Plotting virtual geomagnetic and mean poles

To plot poles we first need a map projection. The matplotlib basemap package does a nice job of that, so let's import it:

```python
from mpl_toolkits.basemap import Basemap
```

The basemap is highly customizable. A view from space type projection ('ortho') is a nice way to view data on a sphere so let's use that one for this example. We will define an object ('m') that is a map projection and then draw additional things on the map such as continents and lat/long lines. Then the VGPs from the study can be plotted using the `IPmag.vgpplot` function.

```python
plt.figure(figsize=(8, 8))
m = Basemap(projection='ortho',lat_0=70,lon_0=230,resolution='c',area_thresh=50000)
m.drawcoastlines(linewidth=0.25)
m.fillcontinents(color='bisque',lake_color='white',zorder=1)
m.drawmapboundary(fill_color='white')
m.drawmeridians(np.arange(0,360,30))
m.drawparallels(np.arange(-90,90,30))

results_Plat = results['vgp_lat'].tolist()
results_Plong = results['vgp_lon'].tolist()

combined_vgps = []
for n in range(0,len(results)):
    if results['vgp_lat'][n]>0:
        IPmag.vgpplot(m,results_Plong[n],results_Plat[n],color='r')
        combined_vgps.append([results_Plong[n],results_Plat[n]])
    if results['vgp_lat'][n]<=0:
        IPmag.vgpplot(m,results_Plong[n]+180,-results_Plat[n],color='b')
        combined_vgps.append([results_Plong[n]+180,-results_Plat[n]])
        
combined_VGP_mean = pmag.fisher_mean(combined_vgps) 

IPmag.poleplot(m,combined_VGP_mean['dec'],combined_VGP_mean['inc'],
               combined_VGP_mean['alpha95'],marker='d')
plt.show()
```
